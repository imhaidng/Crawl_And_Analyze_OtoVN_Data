```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv('data-use.csv')
df
```

```python
print(df.info())
```

```python
# Xóa "km" từ cột 'Số km đã đi'
df['Số km đã đi'] = df['Số km đã đi'].str.replace('km', '')
df
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check for duplicate rows
print(df.duplicated().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates(subset=["Tên xe", "Năm sản xuất"])
```

```python
print(df.duplicated().sum())
```

```python
# Loại bỏ các bản ghi có giá trị "Số km đã đi" bằng giá trị cụ thể
df = df[df["Số km đã đi"] != "HàNội"]
df = df[df["Số km đã đi"] != "Tp.HCM"]
df = df[df["Số km đã đi"] != "BìnhDương"]
df = df[df["Số km đã đi"] != "HảiPhòng"]
df = df[df["Số km đã đi"] != "BìnhPhước"]
df = df[df["Số km đã đi"] != "NinhBình"]
df = df[df["Số km đã đi"] != "BàRịaVũngTàu"]
df = df[df["Số km đã đi"] != "VĩnhPhúc"]
df = df[df["Số km đã đi"] != "ĐàNẵng"]
```

```python
# Chỉ xóa các giá trị NaN trong cột "Tên cột"
df.dropna(subset=["Số km đã đi"], inplace=True)
# Loại bỏ dấu chấm phân tách phần ngàn từ chuỗi
df["Số km đã đi"] = df["Số km đã đi"].str.replace('.', '')

# Chuyển đổi kiểu dữ liệu từ chuỗi sang số nguyên
df["Số km đã đi"] = df["Số km đã đi"].astype(int)
df["Năm sản xuất"] = df["Năm sản xuất"].astype(int)
```

```python
# Impute missing values in numerical columns
df["Giá"] = df["Giá"].fillna(df["Giá"].mean())
df["Số km đã đi"] = df["Số km đã đi"].fillna(df["Số km đã đi"].median())
```

```python
# Kiểm tra giá trị rỗng sau khi xử lý
missing_gia = df["Giá"].isnull().sum()
print("Số lượng giá trị rỗng trong cột 'Tên cột':", missing_gia)

missing_km = df["Số km đã đi"].isnull().sum()
print("Số lượng giá trị rỗng trong cột 'Tên cột':", missing_km)
```
